# Matplotlib 

Creation of graphs

### Importing

```python
# Importing it under some name
import matplotlib.pyplot as plt
```

```python
# Required only in jupyter notebook
%matplotlib inline
```

### The data we want to plot

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# To plot the graph
plt.plot(x,y,'r') # 'r' is the color red
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Graph Title')
plt.show()
```

```python
plt.plot(x,y,'g')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Graph Title')
plt.show()
```

#### Creating multiple graphs

```python
# plt.subplot(nrows,ncols,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
# Create figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])

# Plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel('Set X label')
axes.set_ylabel('Set Y label')
axes.set_title('Title')
```

```python
# Creates blank canvas
fig =  plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # Main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # Inset axes

# Larger figure Axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('X label axes 1')
axes1.set_ylabel('Y label axes 1')
axes1.set_title('Axes 1 Title')

# Insert figure Axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('X label axes 2')
axes2.set_ylabel('Y label axes 2')
axes2.set_title('Axes 2 Title')
```

```python
# Increasing the graph width and height
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x,y,'r')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
```

### Legends and labels

```python
# To add legends in the graph
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="X**2")
ax.plot(x, x**3, label="X**3")
ax.legend()
```
